```python
import pandas as pd

df = pd.read_csv('college_event_feedback_full_dataset.csv')
df.columns = df.columns.str.strip()
df.head()
```

```python
df.columns.tolist()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='rating', palette='coolwarm')
plt.title("Event Rating Distribution")
plt.xlabel("Rating (1 to 5)")
plt.ylabel("Number of Participants")
plt.show()
```

```python
!pip install -q textblob
from textblob import TextBlob
```

```python
df['feedback_sentiment'] = df['feedback'].astype(str).apply(lambda x: TextBlob(x).sentiment.polarity)
df['suggestions_sentiment'] = df['suggestions'].astype(str).apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
df[['participant_name', 'feedback', 'feedback_sentiment', 'suggestions', 'suggestions_sentiment']].head()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['feedback_sentiment'], kde=True, color='green')
plt.title("Feedback Sentiment Distribution")

plt.subplot(1, 2, 2)
sns.histplot(df['suggestions_sentiment'], kde=True, color='orange')
plt.title("Suggestions Sentiment Distribution")

plt.tight_layout()
plt.show()
```

```python
!pip install -q wordcloud
from wordcloud import WordCloud
```

```python
feedback_text = " ".join(df['feedback'].astype(str))
suggestions_text = " ".join(df['suggestions'].astype(str))

WordCloud(width=800, height=400, background_color='white').generate(feedback_text).to_image().show
WordCloud(width=800, height=400, background_color='white').generate(suggestions_text).to_image().show()
```

```python
from wordcloud import WordCloud

feedback_text = " ".join(df['feedback'].astype(str))
suggestion_text = " ".join(df['suggestions'].astype(str))
feedback_wc = WordCloud(width=800, height=400, background_color='white').generate(feedback_text)
plt.imshow(feedback_wc, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud: Feedback")
plt.show()
suggestion_wc = WordCloud(width=800, height=400, background_color='white').generate(suggestion_text)
plt.imshow(suggestion_wc, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud: Suggestions")
plt.show()
```

```python
df.groupby('department')['rating'].mean().sort_values().plot(kind='barh', color='skyblue')
plt.title("Average Rating by Department")
plt.xlabel("Average Rating")
plt.show()
```

```python
sns.countplot(data=df, x='year', palette='Set2')
plt.title("Participation by Year")
plt.show()
```
